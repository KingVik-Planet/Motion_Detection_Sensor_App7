## Different cvt2Color
- RGB to Grayscale: cv2.COLOR_RGB2GRAY
- RGB to HSV (Hue, Saturation, Value): cv2.COLOR_RGB2HSV
- RGB to HLS (Hue, Lightness, Saturation): cv2.COLOR_RGB2HLS
- RGB to YUV (Luma, Blue-difference, Red-difference): cv2.COLOR_RGB2YUV
- RGB to YCrCb (Luma, Chroma blue, Chroma red): cv2.COLOR_RGB2YCrCb
- RGB to LAB (CIE L*a*b* color space): cv2.COLOR_RGB2LAB
- RGB to LUV (CIE L*u*v* color space): cv2.COLOR_RGB2LUV
- RGB to XYZ (CIE 1931 color space): cv2.COLOR_RGB2XYZ
- RGB to BGRA (Adding an alpha channel): cv2.COLOR_RGB2BGRA
- BGRA to RGB (Removing the alpha channel): cv2.COLOR_BGRA2RGB
- HSV to RGB: cv2.COLOR_HSV2RGB
- HLS to RGB: cv2.COLOR_HLS2RGB
- YUV to RGB: cv2.COLOR_YUV2RGB
- YCrCb to RGB: cv2.COLOR_YCrCb2RGB
- LAB to RGB: cv2.COLOR_LAB2RGB
- LUV to RGB: cv2.COLOR_LUV2RGB
- XYZ to RGB:cv2.COLOR_XYZ2RGB
